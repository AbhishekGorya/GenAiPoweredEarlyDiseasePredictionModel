# Data Preprocessing

```python
import pandas as pd
import numpy as np

df = pd.read_csv("multi_disease_dataset_v2.csv")

print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nSample Data:\n", df.head())
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
print("\nAlcohol Missing Values:\n", df['Alcohol'].isnull())
```

```python
duplicates = df.duplicated().sum()
print("Duplicate Rows:", duplicates)

# Remove if any
df = df.drop_duplicates()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numeric Columns:", list(numeric_cols))
print("Categorical Columns:", list(categorical_cols))
```

```python
# Apply realistic physiological limits

df['BMI'] = df['BMI'].clip(10, 45)
df['Systolic_BP'] = df['Systolic_BP'].clip(80, 200)
df['Diastolic_BP'] = df['Diastolic_BP'].clip(50, 130)
df['Cholesterol'] = df['Cholesterol'].clip(100, 350)
df['Creatinine'] = df['Creatinine'].clip(0.5, 3.5)
df['Hemoglobin'] = df['Hemoglobin'].clip(8, 18)
df['Heart_Rate'] = df['Heart_Rate'].clip(40, 180)
df['Daily_Steps'] = df['Daily_Steps'].clip(500, 20000)
df['Sleep_Hours'] = df['Sleep_Hours'].clip(3, 10)
df['Visceral_Fat'] = df['Visceral_Fat'].clip(2, 30)
```

```python
print(df[numeric_cols].describe().T)
```

```python
df['Activity_Level'] = np.where(df['Daily_Steps'] < 4000, 'Low',
                        np.where(df['Daily_Steps'] < 8000, 'Moderate', 'High'))

df['Job_Type'] = np.where(df['Activity_Level'] == 'Low', 'Sedentary', 'Active')
```

```python
print("\nAfter Cleaning:")
print(df.describe(include='all'))
```
