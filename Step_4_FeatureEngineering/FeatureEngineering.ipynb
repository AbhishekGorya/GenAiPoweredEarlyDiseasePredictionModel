```python
import pandas as pd

df = pd.read_csv("final_cleaned_dataset.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['Gender', 'Activity_Level', 'Job_Type', 'Alcohol']

le_dict = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    le_dict[col] = le  # store encoder
```

```python
df.head()
```

```python
df['Pulse_Pressure'] = df['Systolic_BP'] - df['Diastolic_BP']
```

```python
df['Pulse_Pressure'].head()
```

```python
df['MAP'] = (df['Systolic_BP'] + 2 * df['Diastolic_BP']) / 3  #MAP (Mean Arterial Pressure)
```

```python
df['BMI_Category'] = pd.cut(df['BMI'],
                           bins=[0, 18.5, 25, 30, 50],
                           labels=[0, 1, 2, 3])  # Underweight, Normal, Overweight, Obese
```

```python
df['Lifestyle_Risk'] = (
    df['Smoking'] * 2 +
    df['Alcohol'] +
    (df['Activity_Level'] == 0).astype(int) * 2
)
```

# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Exclude target columns
target_cols = [
    'Diabetes_Prob',
    'Hypertension_Prob',
    'Heart_Disease_Prob',
    'Anemia_Prob',
    'Kidney_Disease_Prob'
]

feature_cols = [col for col in numeric_cols if col not in target_cols]

df[feature_cols] = scaler.fit_transform(df[feature_cols])
```

```python
X = df.drop(columns=target_cols)
y = df[target_cols]

print("Feature Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
df.to_csv("feature_engineered_dataset.csv", index=False)
print("✅ Feature engineered dataset saved!")
```
