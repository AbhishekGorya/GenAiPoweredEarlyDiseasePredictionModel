```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_multi_disease_dataset.csv")

sns.set_style("whitegrid")
```

```python
num_features = ['Age', 'BMI', 'Systolic_BP', 'Fasting_Sugar', 'Cholesterol', 'Visceral_Fat']

df[num_features].hist(bins=30, figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
target_cols = [
    'Diabetes_Prob',
    'Hypertension_Prob',
    'Heart_Disease_Prob',
    'Anemia_Prob',
    'Kidney_Disease_Prob'
]

df[target_cols].hist(bins=30, figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x=df['BMI'], y=df['Diabetes_Prob'])
plt.title("BMI vs Diabetes Probability")
plt.show()
```

```python
sns.scatterplot(x=df['Systolic_BP'], y=df['Hypertension_Prob'])
plt.show()

sns.scatterplot(x=df['Cholesterol'], y=df['Heart_Disease_Prob'])
plt.show()

sns.scatterplot(x=df['Hemoglobin'], y=df['Anemia_Prob'])
plt.show()

sns.scatterplot(x=df['Creatinine'], y=df['Kidney_Disease_Prob'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['BMI'], y=df['Diabetes_Prob'])
plt.title("BMI vs Diabetes")
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x=df['Creatinine'], y=df['Kidney_Disease_Prob'])
plt.title("Creatinine vs Kidney Disease")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[num_features])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Activity_Level', y='Diabetes_Prob', data=df)
plt.show()
```
